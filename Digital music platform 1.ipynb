<h2>Why Binary Search Tree (BST)?</h2>
<p>A Binary Search Tree (BST) is chosen for managing the music catalogue due to its efficient search, insertion, and deletion operations, making it ideal for handling large music libraries.</p>
<ol>
    <li><b>Efficient Data Retrieval:</b> Quickly searches for songs based on title, artist, or genre.</li>
    <li><b>Structured Organization of Data:</b> Automatically maintains a sorted order for easy traversal and access.</li>
    <li><b>Dynamic Updates:</b> Supports seamless addition and removal of songs, keeping the catalogue up-to-date.</li>
    <li><b>Handling Large Datasets:</b> Scales well with the growing size of the music library.</li>
    <li><b>Customizable Searches:</b> Enables advanced search options tailored to user preferences.</li>
</ol>

<h2>Why Binary Search Tree (BST) better than Other Data Structures in this project?</h2>
<h5>1. Efficient Searching, Insertion, and Deletion</h5>
<ul>
  <li><b>BST Advantage:</b>
    <p>Operations like searching, inserting, and deleting songs are efficient, with a time complexity of <strong>O(log n)</strong> in a balanced BST.</p>
    <p>These operations ensure that even as the catalogue grows, performance remains optimal.</p>
  </li>
  <li><b>Compared to Arrays:</b>
    <p>Arrays require <strong>O(n)</strong> for insertions and deletions due to the need for shifting elements, making them inefficient for dynamic updates.</p>
  </li>
  <li><b>Compared to Linked Lists:</b>
    <p>Linked lists require <strong>O(n)</strong> for searching, as each element must be traversed sequentially.</p>
  </li>
</ul>
<h5>2. Scalability</h5>

<p>Scalability means how well a system can handle growth. In the context of a project or a data structure like a <b>Binary Search Tree (BST)</b>, scalability refers to its ability to manage increasing amounts of data or users without becoming slow or inefficient.</p>
<p>A <b>Balanced BST</b> is scalable because it keeps operations (like search, insert, or delete) efficient, even with large datasets.</p>
<p>On the other hand, less scalable structures, like an array, would struggle as the library grows due to slower operations like shifting elements.</p>


<h1>How BST Works in  Project</h1>

<h4>1. Structure of the BST</h4>
<ul>
  <li>Each <strong>node</strong> in the BST represents a <strong>song</strong> in the music catalogue.</li>
  <li>The BST is organized based on the <strong>song_id</strong> (a unique identifier for each song).</li>
  <li>The BST follows the <strong>binary search property</strong>:
    <ul>
      <li>All songs in the <strong>left subtree</strong> of a node have a song_id <strong> less than</strong> the node's song_id.</li>
      <li>All songs in the <strong>right subtree</strong> of a node have a song_id<strong> greater than</strong> the node's song_id.</li>
    </ul>
  </li>
</ul>


<h4>2.How Operations Work in the BST</h4>
<p>Here’s how the BST handles key operations in your project:</p>

<h5>a. Insertion</h5>
<ul>
  <li>When a new song is added to the music catalogue:
    <ul>
      <li>Start at the root node.</li>
      <li>Compare the new song's <strong>song_id</strong> with the current node’s <strong>song_id</strong>.</li>
      <li>If the new song’s <strong>song_id</strong> is smaller, move to the left subtree.</li>
      <li>If it’s larger, move to the right subtree.</li>
      <li>Repeat until you find an empty spot, then insert the new song.</li>
    </ul>
  </li>
</ul>

<h5>b. Search</h5>
<ul>
  <li>When a user searches for a song:
    <ul>
      <li>Start at the root node.</li>
      <li>Compare the search <strong>song_id</strong> with the current node’s <strong>song_id</strong>.</li>
      <li>If they match, return the song.</li>
      <li>If the search <strong>song_id</strong> is smaller, move to the left subtree.</li>
      <li>If it’s larger, move to the right subtree.</li>
      <li>Repeat until you find the song or determine it’s not in the tree.</li>
    </ul>
  </li>
</ul>

<h5>c. Deletion</h5>
<ul>
  <li>When a song is removed:
    <ul>
      <li>Locate the node to delete.</li>
      <li>Handle one of these cases:
        <ul>
          <li><b>No children:</b> Simply remove the node.</li>
          <li><b>One child:</b> Replace the node with its child.</li>
          <li><b>Two children:</b> Find the node's <i>in-order successor</i> (smallest node in the right subtree) and replace the node with it.</li>
        </ul>
      </li>
    </ul>
  </li>
</ul>

<h5>d. Traversal</h5>
<ul>
  <li><b>In-order Traversal:</b>
    <p>Visits nodes in ascending order of their <strong>song_id</strong>. Useful for displaying the entire music catalogue in sorted order.</p>
  </li>
</ul>

<h5>e. Minimum and Maximum</h5>
<ul>
  <li><b>Find Minimum:</b>
    <p>Start at the root and keep moving to the left until you reach the leftmost node.</p>
  </li>
  <li><b>Find Maximum:</b>
    <p>Start at the root and keep moving to the right until you reach the rightmost node.</p>
  </li>
</ul>


<h4>Example </h4>
<p>Suppose  music catalogue initially has these songs:</p>
<ol>
  <li>Song A (<strong>song_id = 10</strong>)</li>
  <li>Song B (<strong>song_id = 20</strong>)</li>
  <li>Song C (<strong>song_id = 5</strong>)</li>
</ol>

<h5>Tree Structure</h5>
<p>1. Song A (<strong>song_id = 10</strong>) becomes the root.<br>
2. Song C (<strong>song_id = 5</strong>) goes to the left of Song A because <strong>5 &lt; 10</strong>.<br>
3. Song B (<strong>song_id = 20</strong>) goes to the right of Song A because <strong>20 &gt; 10</strong>.</p>

<p>The tree now looks like this:</p>
<pre>
      10
     /  \
    5    20
</pre>

<h5>Operations Example</h5>
<ul>
  <li><b>Search for Song C:</b>
    <p>Start at the root (10), go left (5), and find Song C.</p>
  </li>
  <li><b>Add Song D (<strong>song_id = 15</strong>):</b>
    <p>Start at 10, go right (20), and place 15 to the left of 20.</p>
  </li>
  <li><b>Delete Song A (<strong>song_id = 10</strong>):</b>
    <p>Replace it with the in-order successor (15).</p>
  </li>
</ul>

<h3>Why Linked List is Preferred for Playlist Management</h3>
<ol>
    <li>
        <b>Dynamic Size:</b> 
        A singly linked list allows the playlist to grow or shrink as songs are added or removed without requiring pre-allocated memory, making it efficient for dynamic updates.
    </li>
    <li>
        <b>Efficient Insertions and Deletions:</b> 
        Adding or removing songs is efficient, especially at the beginning or end of the list (<b>O(1)</cb>), unlike arrays that require shifting elements.
    </li>
    <li>
        <b>Simple Sequential Traversal:</b> 
        A singly linked list supports sequential traversal using the next pointer, aligning well with the way songs are typically played in a playlist.
    </li>
    <li>
        <b>Ease of Implementation:</b> 
        The logic for managing songs (adding, removing, and iterating) is straightforward and fits the requirements of playlist management without unnecessary complexity.
    </li>
</ol>


<h3>Why Not Use Binary Search Tree (BST) for Playlist Management?</h3>
<ol>
    <li>
        <b>Ordered Traversal Overhead:</b> 
        BSTs prioritize sorted data access, which is unnecessary for playlists that typically require sequential playback rather than sorted song orders.
    </li>
    <li>
        <b>Complexity for Basic Operations:</b> 
        Operations like adding or removing songs are more complex in a BST compared to the simplicity of linked lists, especially when maintaining tree balance.
    </li>
    <li>
        <b>Unnecessary Structural Overhead:</b> 
        The hierarchical structure of a BST introduces additional memory overhead for left and right child pointers, which are not required for managing a simple sequential playlist.
    </li>
    <li>
        <b>Dynamic Order Changes:</b> 
        Playlists often involve dynamic reordering (e.g., moving a song up or down), which is more intuitive and efficient with a linked list than restructuring a BST.
    </li>
</ol>
